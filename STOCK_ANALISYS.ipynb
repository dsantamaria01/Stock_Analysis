# Take the data

```python
tech_list = ['AAPL','GOOG','MSFT','AMZN', 'TSLA','META','COKE']

end = datetime.now()
start = datetime(end.year - 2,end.month,end.day)

for stock in tech_list: 
    globals()[stock] = data.get_data_yahoo(stock,start,end)
```
